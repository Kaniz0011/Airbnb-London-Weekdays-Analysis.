'host_is_superhost'

```python
df.columns
```

```python
superhosts=df.groupby('room_type')['host_is_superhost'].mean()*100
print(superhosts)
```

```python
# visualization of Count of Listings by Room Type

sns.countplot(x="room_type",width=0.5,data=df)
plt.title("Count of Listings by Room Type")
plt.show()
```

```python
# visualization of 'host_is_superhost'  by percentage

superhosts.plot(kind ='bar')
plt.title("'host_is_superhost'  by percentage")
plt.ylabel('Percentage')
plt.show()
```

real sum

```python
df.columns
```

```python
# Prices of rooms 
realSum=df.groupby('room_type')['realSum'].mean()*100
realSum
```

```python
# visualization of prices 
realSum.plot(kind='bar')
plt.title('Price Distribution of Airnbn')
plt.xlabel('realSum')
plt.ylabel('Price (realSum)')
plt.show()
```

```python
plt.figure(figsize=(5,3))
sns.histplot(df['realSum'], bins=50, kde=True)
plt.title("Price Distribution of Airbnb Listings")
plt.xlabel("Price (realSum)")
plt.show()
```

```python
# Do Superhosts Charge More?

avg_price_superhost = df.groupby('host_is_superhost')['realSum'].mean()
print(avg_price_superhost)
```

```python
sns.boxplot(x='room_type',y='realSum',data=df)
plt.title("Do Superhosts Charge Higher Prices?")
plt.xticks([0,1], ['Not Superhost','Superhost'])
plt.show()
```

<h2 style="font-size:26px; font-weight:bold; color:blaACck;">
8). Location Analysis.

</h2>

```python
#  Basic Stats for Location Columns

print(df[['dist', 'metro_dist', 'attr_index', 'rest_index']].describe())
```

```python
plt.figure(figsize=(7,5))
sns.scatterplot(x='dist', y='realSum', data=df, alpha=0.5)
plt.title("Price vs Distance from City Center")
plt.xlabel("Distance from City Center (km)")
plt.ylabel("Price (realSum)")
plt.show()

# Expect closer to city center  higher price.
```

```python
#  Price vs Metro Distance

plt.figure(figsize=(7,5))
sns.scatterplot(x='metro_dist', y='realSum', data=df, alpha=0.5)
plt.title("Price vs Distance from Metro Station")
plt.xlabel("Distance from Metro (km)")
plt.ylabel("Price (realSum)")
plt.show()

#Listings closer to metro stations may charge higher.
```

```python
#Attractions Index vs Guest Satisfaction

plt.figure(figsize=(7,5))
sns.scatterplot(x='attr_index', y='guest_satisfaction_overall', data=df, alpha=0.5)
plt.title("Attractions Index vs Guest Satisfaction")
plt.xlabel("Attraction Index")
plt.ylabel("Guest Satisfaction Overall")
plt.show()

# See if guests are happier when staying in more attractive areas.
```

```python
#  Restaurant Index vs Guest Satisfaction

plt.figure(figsize=(7,5))
sns.scatterplot(x='rest_index', y='guest_satisfaction_overall', data=df, alpha=0.5)
plt.title("Restaurant Index vs Guest Satisfaction")
plt.xlabel("Restaurant Index")
plt.ylabel("Guest Satisfaction Overall")
plt.show()

# More restaurants nearby could increase satisfaction.
```

```python
#  Correlation Heatmap (All Location + Price + Ratings)

plt.figure(figsize=(8,6))
sns.heatmap(df[['realSum','dist','metro_dist','attr_index','rest_index','guest_satisfaction_overall']].corr(),
            annot=True, cmap='coolwarm')
plt.title("Correlation Between Price, Location & Satisfaction")
plt.show()

# Quick way to check relationships between price, distance, attractions, restaurants, and satisfaction.
```
